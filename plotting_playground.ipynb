```python
import seaborn as sn
import torch as t
import numpy as np
```

```python
x = np.arange(1024)
projection_mean = t.ones([1024]) 
projection_err = t.rand_like(projection_mean)*0.3
```

```python
ax = sn.lineplot(projection_mean)
ax.fill_between(x, projection_mean - projection_err, projection_mean + projection_err, alpha=0.2)
```
